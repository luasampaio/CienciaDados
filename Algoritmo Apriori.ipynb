```python
import pandas as pd
import numpy as np
p
```

```python
dataset = pd.read_csv('census.csv')
dataset.head()
```

```python
dataset['age'] = pd.cut(dataset['age'], bins=[0, 17, 25, 40, 60, 90], labels = ['Faixa 1', 'Faixa 2', 'Faixa 3', 'Faixa 4', 'Faixa 5'])
```

```python
dataset.head()
```

```python
dataset_apriori = dataset[['age','workclass','education','marital-status','relationship','occupation','sex','native-country','income']]
```

```python
dataset_apriori
```

```python
dataset_apriori = dataset_apriori.sample(1000)
dataset_apriori.shape
```

```python
transacoes = []
for i in range(dataset_apriori.shape[0]):
    transacoes.append([str(dataset_apriori.values[i,j])for j in range(dataset_apriori.shape[1])])
```

```python
len(transacoes)
```

```python
transacoes
```

```python
regras = apriori(transacoes, min_support = 0.3, min_confidence = 0.2)
resultados = list(regras)
```
