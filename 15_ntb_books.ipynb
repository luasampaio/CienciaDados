Detectar automaticamente o separador:
 - Use o módulo csv.Sniffer para identificar o separador correto

```python
import csv
with open('/content/drive/MyDrive/.Dataset/books.csv', 'r', encoding='UTF-8') as file:
    dialect = csv.Sniffer().sniff(file.read(1024))
    file.seek(0)
    df2 = pd.read_csv(file, sep=dialect.delimiter)
```

```python
df2.head()
```
